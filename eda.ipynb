```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("datasets/winequality.csv", index_col=0)
```

```python
df
```

```python
df.describe().to_csv("datasets/describe.csv")
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
corr_matrix_fig = plt.gcf()
plt.show()
corr_matrix_fig.savefig("plots\corr_matrix.png")
```

```python
fig, axes = plt.subplots(3, 4, figsize=(12, 7)) 

axes = axes.flatten()

for i, column in enumerate(df.columns[:-1]): 
    sns.histplot(data=df, x=column, kde=(column != "white"), ax=axes[i], bins=30)
    axes[i].set_title(f'Distribution of {column}')

plt.tight_layout()
feature_distribution_fig = plt.gcf()
plt.show()
feature_distribution_fig.savefig("plots/feature_distribution.png")
```

```python
sns.countplot(x='quality', data=df)
plt.title("Distribution of quality")
quality_distribution_fig = plt.gcf()
plt.show()
quality_distribution_fig.savefig("plots/quality_distribution.png")
```
